```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Method 1 (using np.random for simulation)

```python
# Method 1 (using np.random as an output itself)

num_sim = 1000000
```

```python
input_A = np.random.uniform(1,5,num_sim)
input_B = np.random.uniform(2,6,num_sim)
```

```python
equation = input_A + input_B
```

```python
plt.hist(equation, density = True)
plt.axvline(9, color = 'r')
print((equation > 9).sum()/num_sim)
```

Method 2 (using for loop for simulation)

```python
# Method 2 (using for loop with append as the simulation method, describe measure as the output, np.random as an input)

mean = 1
std_dev = 0.1
num_input = 5
num_simulations = 1000
```

```python
all_stats = []

for i in range(num_simulations):

    sales_target = np.random.choice([7500, 10000, 20000, 30000], num_input, p=[0.3, 0.4, 0.2, 0.1])
    pct_to_target = np.random.normal(mean, std_dev, num_input).round(2)

    df = pd.DataFrame(index=range(num_input), data={'Pct_To_Target': pct_to_target,
                                                   'Sales_Target': sales_target})

    df['Sales'] = df['Pct_To_Target'] * df['Sales_Target']

    all_stats.append([df['Sales'].sum().round(0)])
```

```python
results_df = pd.DataFrame(data=all_stats)
results_df.describe()
```
